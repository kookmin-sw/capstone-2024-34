```python
import pickle
```

```python
with open('./res/iot_signature.pkl', 'rb') as f:
    signature = pickle.load(f)
    
with open('./res/iot_stopword.pkl', 'rb') as f:
    stopword = pickle.load(f)
```

```python
print('SIGNAURE')
print(signature)
    
print('STOPWORD')
for i in stopword:
    print(i)
```

```python
filters = [i[0] for i in stopword if i[1] > 99]
```

```python
for cluster, sig in signature.items():
    for token in sig:
        if token[0] not in filters:
            print(token)
```
